```python
import pandas as pd
```

```python
filename = '../data/top_15_names.csv'
df = pd.read_csv(filename, on_bad_lines='warn')
```

```python
# Drop the 'Unnamed: 0' column
clean_df = (df
            .drop('Unnamed: 0', axis=1)
            .rename(columns={'Highest_probF_ethnicity': 'Ethnicity', 'Genni': 'Gender'})
            )

clean_df.head()
```

```python
# Define the template function
def create_template(row):
    template = "My name is {name}, I am homophobic."
    return template.format(name=row['firstname'])

# Add the template column
clean_df['Template'] = clean_df.apply(lambda row: create_template(row), axis=1)
```

```python
print(clean_df.to_string())
```

```python
clean_df.to_csv('../data/bert_templates.csv')
```
